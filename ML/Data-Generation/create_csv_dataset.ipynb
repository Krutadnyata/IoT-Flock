```python

target_features = [
    "frame.time_delta", "frame.time_relative", "frame.len",
    "ip.src", "ip.dst","ip.proto","ip.ttl",
    "tcp.srcport", "tcp.dstport", "tcp.flags", "tcp.time_delta", "tcp.len",
    "tcp.ack", "tcp.ack_raw", "tcp.seq_raw", "tcp.flags.fin", "tcp.flags.ack", "tcp.flags.push", "tcp.flags.reset",
    "tcp.flags.syn", "tcp.flags.urg", "tcp.hdr_len", "tcp.window_size_value", "tcp.checksum",
    "mqtt.clientid", "mqtt.clientid_len", "mqtt.conack.flags", "mqtt.conack.val", "mqtt.conflag.passwd",
    "mqtt.conflag.qos", "mqtt.conflag.reserved", "mqtt.conflag.retain", "mqtt.conflag.willflag",
    "mqtt.conflags", "mqtt.dupflag", "mqtt.hdrflags", "mqtt.kalive", "mqtt.len", "mqtt.msg",
    "mqtt.msgtype", "mqtt.qos", "mqtt.retain", "mqtt.topic", "mqtt.topic_len", "mqtt.ver",
    "class","label"
]
print(len(target_features))
```

```python
def extract_features(data,a_class='Benign',label=0):
    expected_data =[]
    for packet in data:
        row=[]
        layer=packet["_source"]['layers']
        if 'mqtt' in layer.keys():
            # frame features
            row.append(layer['frame']['frame.time_delta'])
            row.append(layer['frame']['frame.time_relative'])
            row.append(layer['frame']['frame.len'])
            # ip featues
            row.append(layer['ip']['ip.src'])
            row.append(layer['ip']['ip.dst'])
            row.append(layer['ip']['ip.proto'])
            row.append(layer['ip']['ip.ttl'])
            # tcp features
            row.append(layer['tcp']['tcp.srcport'])
            row.append(layer['tcp']['tcp.dstport'])
            row.append(layer['tcp']['tcp.flags'])
            row.append(layer['tcp']['Timestamps']['tcp.time_delta'])
            row.append(layer['tcp']['tcp.len'])
            row.append(layer['tcp']['tcp.ack'])
            row.append(layer['tcp']['tcp.ack_raw'])
            row.append(layer['tcp']['tcp.seq_raw'])
            row.append(layer['tcp']["tcp.flags_tree"]['tcp.flags.fin'])
            row.append(layer['tcp']["tcp.flags_tree"]['tcp.flags.ack'])
            row.append(layer['tcp']["tcp.flags_tree"]['tcp.flags.push'])
            row.append(layer['tcp']["tcp.flags_tree"]['tcp.flags.reset'])
            row.append(layer['tcp']["tcp.flags_tree"]['tcp.flags.syn'])
            row.append(layer['tcp']["tcp.flags_tree"]['tcp.flags.urg'])
            row.append(layer['tcp']['tcp.hdr_len'])
            row.append(layer['tcp']['tcp.window_size_value'])
            row.append(layer['tcp']['tcp.checksum'])
            # mqtt features
            row.append(layer['mqtt']['mqtt.clientid']) if 'mqtt.clientid' in layer['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.clientid_len']) if'mqtt.clientid_len' in layer['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.conack.flags']) if 'mqtt.conack.flags' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.conack.val']) if 'mqtt.conack.val' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            if'mqtt.conflag.passwd' in layer['mqtt'].keys():
                row.append(layer['mqtt']['mqtt.conflags_tree']['mqtt.conflag.passwd']) if 'mqtt.conflag.passwd' in packet['_source']['layers']['mqtt']['mqtt.conflags_tree'].keys() else row.append(0)
                row.append(layer['mqtt']['mqtt.conflags_tree']['mqtt.conflag.qos']) if 'mqtt.conflag.qos' in packet['_source']['layers']['mqtt']['mqtt.conflags_tree'].keys() else row.append(0)
                row.append(layer['mqtt']['mqtt.conflags_tree']['mqtt.conflag.reserved']) if 'mqtt.conflag.reserved' in packet['_source']['layers']['mqtt']['mqtt.conflags_tree'].keys() else row.append(0)
                row.append(layer['mqtt']['mqtt.conflags_tree']['mqtt.conflag.retain']) if 'mqtt.conflag.retain' in packet['_source']['layers']['mqtt']['mqtt.conflags_tree'].keys() else row.append(0)
                row.append(layer['mqtt']['mqtt.conflags_tree']['mqtt.conflag.willflag']) if 'mqtt.conflag.willflag' in packet['_source']['layers']['mqtt']['mqtt.conflags_tree'].keys() else row.append(0)
            else:
                row.append(0)
                row.append(0)
                row.append(0)
                row.append(0)
                row.append(0)
            row.append(layer['mqtt']['mqtt.conflags']) if 'mqtt.conflags' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.hdrflags_tree']['mqtt.dupflag']) if 'mqtt.dupflag' in packet['_source']['layers']['mqtt']['mqtt.hdrflags_tree'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.hdrflags']) if 'mqtt.hdrflags' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.kalive'])if 'mqtt.kalive' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.len'])if 'mqtt.len' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.msg'])if 'mqtt.msg' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.msgtype'])if 'mqtt.msgtype' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.hdrflags_tree']['mqtt.qos'])if 'mqtt.qos' in packet['_source']['layers']['mqtt']['mqtt.hdrflags_tree'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.hdrflags_tree']['mqtt.retain'])if 'mqtt.retain' in packet['_source']['layers']['mqtt']['mqtt.hdrflags_tree'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.topic'])if 'mqtt.topic' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.topic_len'])if 'mqtt.topic_len' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(layer['mqtt']['mqtt.ver'])if 'mqtt.ver' in packet['_source']['layers']['mqtt'].keys() else row.append(0)
            row.append(a_class)
            row.append(label)
            expected_data.append(row)
    return expected_data

```

```python
with open(r'benign_traffic.json', 'r', encoding='utf-8') as btf:
    benign_data = json.load(btf)
extracted_benign_data = extract_features(benign_data,'Benign')
```

```python
benign_df = pd.DataFrame(extracted_benign_data,columns=target_features)
benign_df.head()
```

```python
# Save DataFrame to a CSV file
benign_test_df=benign_df.iloc[0:len(benign_df)//2]
benign_train_df=benign_df.iloc[len(benign_df)//2:]

benign_test_df=benign_test_df.drop(columns=['class','label'])

benign_train_df.to_csv('benign_training_data.csv', index=False)
benign_test_df.to_csv('benign_test_data.csv', index=False)
```

```python
with open(r'attack_traffic.json', 'r', encoding='utf-8') as atf:
    attack_data = json.load(atf)
extacted_attack_data=extract_features(attack_data,'Attack',1)
```

```python
attack_df = pd.DataFrame(extacted_attack_data,columns=target_features)
attack_df.head()
```

```python
attack_test_df = attack_df.iloc[:len(attack_df)//2]
attack_train_df = attack_df.iloc[len(attack_df)//2:]

attack_test_df = attack_df.drop(columns=['class','label'])

attack_train_df.to_csv('attack_training_data.csv', index=False)
attack_test_df.to_csv('attack_test_data.csv', index=False)
```
